```
import matplotlib.pyplot as plt
import numpy as np
```

```
import seaborn as sns
sns.set()
```

```
fig =  plt.figure()
fig.show()
```

```
plt.plot([0,1, 8], [1,3, 4], label='line', color='green')
plt.xlabel('axis X')
plt.ylabel('axis Y')
plt.title('MY PLOT')
_ = plt.legend()
```

```
x= np.arange(-3, 5, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)
plt.plot(x, x_sin, color='red')
plt.plot(x, x_cos, 'b--')
```

```
plt.subplot(121)
plt.plot(x, x_sin)
plt.subplot(122)
plt.plot(x, x_cos)
```

```
fig = plt.figure(figsize=(12, 6))
plt.subplot(131)
plt.plot(x, x_sin)
plt.subplot(132)
plt.plot(x, x_cos, 'r--')
plt.subplot(133)
plt.plot(x, x_cos, 'go')
```

```
plt.bar(x=[1,2,3,4,5], height=[2,3,4.3,1,6])
```

```
plt.barh(y=[1,2,3,4,5], width=[2,3,4.3,1,6], color='green')
```
