```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
```

```
df = sns.load_dataset('tips')
df.head()
```

```
df.info()
```

```
df.describe().T
```

```
df.describe(include=['category']).T
```

```
sns.set(font_scale=1.2)
_ = sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size', col='time', row='smoker')
```

```
sns.catplot(data=df, x='day', y='total_bill', col='time', kind='swarm')
```

```
sns.catplot(data=df, x='day', y='total_bill', col='time', kind='box')
```

```
sns.catplot(data=df, x='day', y='total_bill', col='time', kind='violin')
```

```
sns.catplot(data=df, x='day', y='total_bill', col='time', kind='bar')
```

```
sns.catplot(data=df, x='day', kind='count')
```
